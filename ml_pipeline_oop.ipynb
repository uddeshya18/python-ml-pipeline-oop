```python
class DataLoader:
  def load_data(self):
    print("Loading data....")
    data = [1,2,3,4,5]
    return data

class SimpleModel:
  def train(self,data):
    print("Training model with data:", data)
    self.mean = sum(data)/ len(data)

  def predict(self,x):
    return x * self.mean
```

```python
def run_pipeline():
  loader = DataLoader()
  data = loader.load_data()

  model = SimpleModel()
  model.train(data)

  prediction = model.predict(10)
  print("Prediction:", prediction)

run_pipeline()
```
